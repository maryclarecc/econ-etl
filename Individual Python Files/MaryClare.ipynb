```python
import pandas as pd
```

```python
college = pd.read_csv('Data/college_prices.csv')
salary = pd.read_csv('Data/median_income.csv')
college
```

```python
#keeping only the first year of each in range
college["Year"] = college["Year"].str.split('-').str[0]
college["Year"] = pd.to_datetime(college.Year, format='%Y')
#turning each year into months
college = college.set_index('Year').resample('MS').ffill().reset_index()
college = college.rename(columns={'Year': 'Date', 'Public 4-Year': 'Public', 'Private 4-Year': 'Private'})
college
```

```python
#converting to datetime
salary["DATE"] = pd.to_datetime(salary.DATE)
#turning each year into months
salary = salary.set_index('DATE').resample('MS').ffill().reset_index()
salary = salary.rename(columns={'DATE': 'Date', 'MEHOINUSA672N': 'Salary'})
salary
```
